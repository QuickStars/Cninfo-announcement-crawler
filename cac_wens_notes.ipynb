```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import calendar
import datetime
import re
import os
import sys
from tqdm import tqdm
```

```python
def get_cninfo_notes(url, downloadDir, ts = 5):
    chrome_options = webdriver.ChromeOptions()
    prefs = {'profile.default_content_settings.popups': 0, #防止保存弹窗
        'download.default_directory':downloadDir,#设置默认下载路径
        'profile.default_content_setting_values.automatic_downloads':1 #允许多文件下载
        }
    chrome_options.add_experimental_option('prefs', prefs)
    
    browser = webdriver.Chrome(options=chrome_options)
    browser.get(url)
    
    data = browser.page_source
    pageNum = int(re.findall('<ul class="el-pager">.*>(.*?)</li></ul><button type',data)[0])
    
    datas = []
    datas.append(data)
    for i in range(1,pageNum):
        browser.find_element_by_xpath('//*[@id="main"]/div[3]/div/div[2]/div/div/div[2]/div[1]/div[3]/div/div/button[2]').click()
        time.sleep(ts)
        data = browser.page_source
        datas.append(data)
    allData = ''.join(datas)
    
    p_title = '<div class="cell"><a target="_blank" href=".*?" data-id=".*?" data-orgid=".*?" data-seccode=".*?" class=.*?><span title="(.*?)" class="r-title">.*?</div>'
    p_href = '<div class="cell"><a target="_blank" href="(.*?)" data-id=".*?" data-orgid=".*?" data-seccode=".*?" class=.*?><span title=".*?" class="r-title">.*?</div>'
    p_date = '<div class="cell"><a target="_blank" href=".*?" data-id=".*?" data-orgid=".*?" data-seccode=".*?" class=.*?><span title=".*?" class="r-title">.*?<span class="time">\n\s+(.*?)\n'
    p_bulletId = '<div class="cell"><a target="_blank" href=".*?" data-id="(.*?)" data-orgid=".*?" data-seccode=".*?" class=.*?><span title=".*?" class="r-title">.*?</div>'
    p_fileType = '<div class="cell"><a target="_blank" href=".*?" data-id=".*?" data-orgid=".*?" data-seccode=".*?" class=.*?><span title=".*?" class="r-title">.*?<i class="iconfont iconnotice-(.*?)">'

    
    title = re.findall(p_title,allData)
    href = re.findall(p_href,allData)
    bulletId = re.findall(p_bulletId,allData)
    date = re.findall(p_date,allData)
    fileType = re.findall(p_fileType,allData)
    
    downloadUrls = []
    for i in range(len(href)):
        href[i] = href[i].replace('%2F','/')
        href[i] = href[i].replace('%3A',':')
        if fileType[i] == 'DOC1':
            downloadUrl = re.findall('src=(.*?)$',href[i])[0]
        elif fileType[i] == 'EXL1':
            downloadUrl = re.findall('src=(.*?)$',href[i])[0]
        else:
            downloadUrl = 'http://www.cninfo.com.cn/new/announcement/download?bulletinId='+bulletId[i]+'&announceTime='+date[i]
        downloadUrls.append(downloadUrl)
    
    for i in tqdm(range(len(downloadUrls))):
        browser.get(downloadUrls[i])
        time.sleep(1)
        sys.stdout.flush()
    
    browser.close()
    
```

```python
url = 'http://www.cninfo.com.cn/new/disclosure/stock?stockCode=300498&orgId=9900009247#research'
downloadDir = os.getcwd()+"\\announcements\\wens\\notes"
get_cninfo_notes(url, downloadDir)
```
